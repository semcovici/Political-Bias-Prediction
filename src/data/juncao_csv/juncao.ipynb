```python
import os
```

```python
import numpy as np
import pandas as pd
```

```python
df_final = pd.DataFrame(columns = ["Partido", "Conteudo", "URL", "Vies"])
```

```python
path = "../../../dataset/raw/scrapping/"
for file in os.listdir(path):
    add = []
    if file[-3:] == "csv":
        df_aux = pd.read_csv(path + file)
        print(file)

        if "psb.csv" in file:
            df_aux["Partido"] = "PSB"
            df_aux["Conteudo"] = df_aux["Título"] + " | " + df_aux["Data"] + " | " + df_aux["Conteudo"]
            df_aux["Vies"] = "centro"

        if "pt" in file:
            df_aux["Partido"] = "PT"
            df_aux["Conteudo"] = df_aux["Título"] + " | " + df_aux["Data"] + " | " + df_aux["Conteudo"]
            df_aux["Vies"] = "esquerda"

        if "uniaoBrasil" in file:
            df_aux["Partido"] = "União Brasil"
            df_aux["Conteudo"] = df_aux["Título"] + " | " + df_aux["Data"] + " | " + df_aux["Conteudo"]
            df_aux["Vies"] = "direita"

        if "pcDoB" in file:
            df_aux["Partido"] = "PCDoB"
            df_aux["Conteudo"] = df_aux["Título"] + " | " + df_aux["Data"] + " | " + df_aux["Conteudo"]
            df_aux["Vies"] = "esquerda"

        if "rede" in file:
            df_aux["Partido"] = "Rede"
            df_aux["Conteudo"] = df_aux["title"] + " | " + df_aux["date"] + " | " + df_aux["content"]
            df_aux["URL"] = df_aux["url"]
            df_aux["Vies"] = "esquerda"

        if "mdb" in file:
            df_aux["Partido"] = "MDB"
            df_aux["Conteudo"] = df_aux["title"] + " | " + df_aux["date"] + " | " + df_aux["content"]
            df_aux["URL"] = df_aux["url"]
            df_aux["Vies"] = "centro"

        if "pv" in file:
            df_aux["Partido"] = "PV"
            df_aux["Conteudo"] = df_aux["title"] + " | " + df_aux["date"] + " | " + df_aux["content"]
            df_aux["URL"] = df_aux["url"]
            df_aux["Vies"] = "centro"

        if "psb_output" in file:
            df_aux["Partido"] = "PSB"
            df_aux["Conteudo"] = df_aux["title"] + " | " + df_aux["date"] + " | " + df_aux["content"]
            df_aux["URL"] = df_aux["url"]
            df_aux["Vies"] = "centro"

        if "NOVO" in file:
            df_aux["Partido"] = "Novo"
            df_aux["Conteudo"] = df_aux["artigo"]
            df_aux["URL"] = df_aux["url"]
            df_aux["Vies"] = df_aux["label"]

        if "PCB" in file:
            df_aux["Partido"] = "PCB"
            df_aux["Conteudo"] = df_aux["artigo"]
            df_aux["URL"] = df_aux["url"]
            df_aux["Vies"] = df_aux["label"]

        if "PDT" in file:
            df_aux["Partido"] = "PDT"
            df_aux["Conteudo"] = df_aux["artigo"]
            df_aux["URL"] = df_aux["url"]
            df_aux["Vies"] = df_aux["label"]

        if "PL" in file:
            df_aux["Partido"] = "PL"
            df_aux["Conteudo"] = df_aux["artigo"]
            df_aux["URL"] = df_aux["url"]
            df_aux["Vies"] = df_aux["label"]

        if "PP" in file:
            df_aux["Partido"] = "PP"
            df_aux["Conteudo"] = df_aux["artigo"]
            df_aux["URL"] = df_aux["url"]
            df_aux["Vies"] = df_aux["label"]

        if "PSTU" in file:
            df_aux["Partido"] = "PSTU"
            df_aux["Conteudo"] = df_aux["artigo"]
            df_aux["URL"] = df_aux["url"]
            df_aux["Vies"] = df_aux["label"]

        if "PSOL" in file:
            df_aux["Partido"] = "PSOL"
            df_aux["Conteudo"] = df_aux["artigo"]
            df_aux["URL"] = df_aux["url"]
            df_aux["Vies"] = df_aux["label"]

        df_final = pd.concat([df_final, df_aux[["Partido", "Conteudo", "URL", "Vies"]]])
```

```python
df_final = df_final.reset_index().drop(columns = "index")
```

```python
df_final = df_final.dropna()
df_final_nan = df_final.copy()
```
